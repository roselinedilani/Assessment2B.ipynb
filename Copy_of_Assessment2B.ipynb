###### Department wise codekata performence (pie chart)

```python
import matplotlib.pyplot as plt
import pandas as pd
df =  pd.read_csv('college_1.csv')
Department_data = df["Department"]
Score_data = df["CodeKata Score"]
colors = ['red', 'pink', 'steelblue']
df.groupby(['Department']).mean().plot(kind='pie', y='CodeKata Score',autopct='%1.0f%%',colors=colors)
plt.title("Department wise codekata performence")
plt.show()
```
